```python
import pandas as pd
import numpy as np
import random
```

```python
def descriptive_stats(df, column):
    s = pd.to_numeric(df[column], errors="coerce").dropna()
    stats = {
        "count": s.count(),
        "mean": s.mean(),
        "median": s.median(),
        "mode": s.mode()[0] if not s.mode().empty else np.nan,
        "std_dev": s.std(),
        "min": s.min(),
        "Q1": s.quantile(0.25),
        "Q3": s.quantile(0.75),
        "max": s.max(),
        "range": s.max() - s.min(),
        "IQR": s.quantile(0.75) - s.quantile(0.25),
        "skewness": (s.mean() - s.median()) / s.std() if s.std() else np.nan
    }
    return stats
```

```python
df = pd.read_csv("Student Mental health.csv")
```

```python
dep_var = "Do you have Anxiety?"
```

```python
df[dep_var] = df[dep_var].replace({"Yes": 1, "No": 0, "yes": 1, "no": 0}).infer_objects(copy=False)
```

```python
sample_df = df.sample(n=min(150, len(df)), random_state=42)
```

```python
results = descriptive_stats(sample_df, dep_var)
```

```python
print("Random Sample Size:", len(sample_df))
print(f"\nDescriptive Statistics for Dependent Variable: {dep_var}\n")
for key, value in results.items():
    print(f"{key:10}: {value}")
```
