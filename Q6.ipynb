```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("Student Mental health.csv", sep=",", engine="python", on_bad_lines="skip")
```

```python
dep_var = "Age"
```

```python
df[dep_var] = pd.to_numeric(df[dep_var], errors="coerce")
```

```python
df = df.dropna(subset=[dep_var])
```

```python
sns.set(style="whitegrid")
```

```python
plt.figure(figsize=(6, 4))
plt.scatter(df.index, df[dep_var], color='teal', alpha=0.7)
plt.title("Scatter Plot of Students' Ages", fontsize=14)
plt.xlabel("Respondent Index", fontsize=12)
plt.ylabel("Age", fontsize=12)
plt.savefig("scatter_plot_age.png", dpi=300, bbox_inches="tight")
plt.show()
```

```python
plt.figure(figsize=(4, 5))
sns.boxplot(y=df[dep_var], color='skyblue')
plt.title("Box Plot of Students' Ages", fontsize=14)
plt.ylabel("Age", fontsize=12)
plt.savefig("box_plot_age.png", dpi=300, bbox_inches="tight")
plt.show()
```

```python
plt.figure(figsize=(6, 4))
plt.hist(df[dep_var], bins=6, color='coral', edgecolor='black')
plt.title("Histogram of Students' Ages", fontsize=14)
plt.xlabel("Age", fontsize=12)
plt.ylabel("Frequency", fontsize=12)
plt.savefig("histogram_age.png", dpi=300, bbox_inches="tight")
plt.show()
```

```python
plt.figure(figsize=(4, 3))
sns.heatmap(df[[dep_var]].corr(), annot=True, cmap="YlGnBu", linewidths=1)
plt.title("Heat Map: Correlation of Age Variable", fontsize=14)
plt.savefig("heatmap_age.png", dpi=300, bbox_inches="tight")
plt.show()
```
