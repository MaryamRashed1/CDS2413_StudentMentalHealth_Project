```python
import pandas as pd
from scipy.stats import pearsonr, spearmanr
```

```python
df = pd.read_csv("Student Mental health.csv", sep=",", engine="python", on_bad_lines="skip")
```

```python
df.columns = df.columns.str.strip()
```

```python
df["Do you have Anxiety?"] = df["Do you have Anxiety?"].astype(str).str.strip().str.lower().map({
    "yes": 1,
    "no": 0
})
```

```python
independent_var = "Do you have Anxiety?"
dependent_var = "Age"
```

```python
df_clean = df[[independent_var, dependent_var]].dropna()
```

```python
pearson_corr, pearson_p = pearsonr(df_clean[independent_var], df_clean[dependent_var])
spearman_corr, spearman_p = spearmanr(df_clean[independent_var], df_clean[dependent_var])
```

```python
print("Pearson Correlation:", round(pearson_corr, 3))
print("P-value (Pearson):", round(pearson_p, 3))
print("Spearman Correlation:", round(spearman_corr, 3))
print("P-value (Spearman):", round(spearman_p, 3))
```
